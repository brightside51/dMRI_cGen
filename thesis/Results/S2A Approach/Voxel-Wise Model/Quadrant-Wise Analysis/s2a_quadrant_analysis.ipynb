## Possible Quadrant Divisions

*Quadrant Size* ---> *Number of Divisions*

4 x 4 -------------> 20 x 23 = 460

8 x 23 ------------> 10 x 4 = 40

16 x 23 -----------> 5 x 4 = 20

```python
# Result Settings
version = 3
metric = 'ssim'
LENGTH_H = 80
LENGTH_V = 92
quadrant_h = 16
quadrant_v = 23

# Result Data Reading
data = np.loadtxt(f"Raw Data/s2a_voxel_v{version}_quadrant_{quadrant_h}x{quadrant_v}_{metric}.txt")
data = data.reshape((LENGTH_H // quadrant_h, LENGTH_V // quadrant_v))
if metric == 'ssim': data = 1 - data
sns.heatmap(data, square = True, cmap = 'cividis',
            xticklabels = False, yticklabels = False,)
```
